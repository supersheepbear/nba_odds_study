```python
import pandas as pd
from datetime import datetime
```

```python
odds_f = pd.read_csv(r"D:\project\nba_crawler\machine_learning\odds.csv")
predict_f = pd.read_csv(r"D:\project\nba_crawler\machine_learning\predictions.csv")
```

```python
def get_datetime_odds_f(date_string):
    month = int(date_string[8:10])
    if month >= 10:
        year = int(date_string[0:4])
    else:
        year = int("20" + date_string[5:7])
    day = int(date_string[-2:])
    return datetime(year, month, day)
        
```

```python
#odds_f.loc[:, "datetime"] = odds_f.index
odds_f.loc[:, "datetime"]  = odds_f.loc[:, "Date"].apply(get_datetime_odds_f)
odds_f.loc[:, "key"] = odds_f["datetime"].astype(str) + odds_f["Home"] + odds_f["Away"]
```

```python
def get_datetime_predict_f(date_string):
    str_list = date_string.split("-")
    month = str_list[0]
    if len(str_list[1]) == 1:
        day = "0" + str_list[1]
    else:
        day = str_list[1]
    
    if int(month) >= 10:
        year = str_list[2]
    else:
        year = int("20" + str_list[3])
    
    return datetime(int(year), int(month), int(day))
        
```

```python
predict_f.loc[:, "datetime"]  = predict_f.loc[:, "Date"].apply(get_datetime_predict_f)
predict_f.loc[:, "key"] = predict_f["datetime"].astype(str) + predict_f["TEAM_NAME"] + predict_f["TEAM_NAME.1"]
```

```python
predict_f.set_index("key", drop=True, inplace=True)
odds_f.set_index("key", drop=True, inplace=True)
```

```python
combine_f = predict_f.merge(odds_f, how="inner", left_index=True, right_index=True)
```
