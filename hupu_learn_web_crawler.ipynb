```python
import requests
from pyquery import PyQuery as pq
from fake_useragent import UserAgent 
```

```python
url = 'https://bbs.hupu.com/rockets'
r = requests.get(url)
print(r)
```

```python
print(r.text)
```

```python
#初始化，先让pyquery认识它，才能解析它
doc = pq(r.text)
```

```python
#class属性为for-list的ul标签 下面的li标签
#因为我们要分别取出每个li标签的内容，所以要加上items方法方便遍历
info = doc('div.bbs-sl-web-holder li.bbs-sl-web-post-body').items()#
```

```python
#对每个li标签，提取它嵌套的class属性为truetit的a标签的内容
#获取文本用text方法
for a, i in enumerate(info):
    #print(a)
    title = i('a.p-title').text()
    print(title)
```

```python
def get_title(n):
    url = f'https://bbs.hupu.com/rockets-{n}'
    r = requests.get(url)
    doc = pq(r.text)
    info = doc('div.bbs-sl-web-holder li.bbs-sl-web-post-body').items()
    for i in info:
            title = i('a.p-title').text()
            print(title)
```

```python
for i in range(11):
    get_title(i)
```
